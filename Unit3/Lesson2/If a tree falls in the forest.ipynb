```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn import ensemble
from IPython.display import Image
from sklearn.model_selection import cross_val_score

import time

import pydotplus
import graphviz

%matplotlib inline
```

```python
# Replace the path with the correct path for your data.
y2015 = pd.read_csv(
    'https://www.dropbox.com/s/0so14yudedjmm5m/LoanStats3d.csv?dl=1',
    skipinitialspace=True,
    header=1
)

# Note the warning about dtypes.
```

```python
y2015.head()
```

```python
categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
# Convert ID and Interest Rate to numeric.
y2015['id'] = pd.to_numeric(y2015['id'], errors='coerce')
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

# Drop other columns with many unique variables
y2015.drop(['url', 'emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state', 'desc'], 1, inplace=True)
```

```python
y2015.tail()
```

```python
y2015 = y2015[:-2]
y2015.drop(['id','member_id','issue_d'],1, inplace=True)
y2015.dropna(axis=1,inplace=True)

y2015_object = y2015.select_dtypes(include='object')
y2015_float = y2015.select_dtypes(include='float64')

y2015_object_dummies = pd.get_dummies(y2015_object.drop('loan_status',1))
y2015_new = pd.concat([y2015_float,y2015_object_dummies],axis=1,join='outer')
```

```python
#y2015_object_dummies.info()
```

Decision Tree

```python
start_time = time.clock()

X_tree = y2015_new
Y_tree = y2015['loan_status']

decision_tree = tree.DecisionTreeClassifier(criterion='entropy',max_features=1,max_depth=4,random_state = 1337)
decision_tree.fit(X_tree, Y_tree)

print(cross_val_score(decision_tree, X_tree, Y_tree, cv=10))

print("--- %s seconds ---" % (time.clock() - start_time))
```

Random Forest

```python
start_time = time.clock()

rfc = ensemble.RandomForestClassifier()
X_rf = y2015_new
Y_rf = y2015['loan_status']

print(cross_val_score(rfc, X_rf, Y_rf, cv=10))

print("--- %s seconds ---" % (time.clock() - start_time))
```
