```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

This data set is from the United Nations databse. This data set contains total municiple waste collected in Japan and South Korea from 1990 - 2014. This is a relatively simple data set. First we read the csv file. In this csv file, since both data have the same year, so I only read the year once. For collected waste of Japan and South Korea, I read these data using iloc arguement and store them in two different variables. Japan_Waste and Korea_Waste. 

```python
df = pd.read_csv('UNdata_Export_20181026_084649514.csv')

Year = df.iloc[0:25,1]

Japan_Waste = df.iloc[0:25,2]

Korea_Waste = df.iloc[25:50,2]
```

First I plot a line plot for the Japan data. X axis is year from 1990 - 2014, and Y axis is the municiple waste collected each year, unit in x1000 Metric ton. According to the data, the total municiple waste collected had rised from 1990 - 2000. However, it started to decrease since 2000. 

```python
plt.figure(figsize=(15,8))
plt.plot(Year, Japan_Waste,color = 'red')
plt.xlabel("Year")
plt.ylabel("x1000 Metric Ton")
plt.title("Total Municiple Waste Collected - Japan")
plt.show()
```

Second we plot the South Korea total municiple waste collected from 1990 - 2014. We can see that the total collected waste peak at 1991 and then started to decrease. 

```python
plt.figure(figsize=(15,8))
plt.plot(Year, Korea_Waste,color = 'green')
plt.xlabel("Year")
plt.ylabel("x1000 Metric Ton")
plt.title("Total Municiple Waste Collected - South Korea")
plt.show()
```

Now we compare the total collected waste between Japan and South Korea. From the chart below, we can see that the total collected waste in South Korea is much lesser compare to Japan data. 

```python
plt.figure(figsize=(15,8))
plt.plot(Year, Japan_Waste,color = 'red', label='Japan')
plt.plot(Year, Korea_Waste, color = 'green', label='South Korea')
plt.legend(loc="upper left")
plt.xlabel("Year")
plt.ylabel("x1000 Metric Ton")
plt.title("Total Municiple Waste Collected - Japan vs. South Korea")
plt.show()
```

Using the same data, we look at it at a scatter plot. Both Japan and South Korea data are in the same plot. 

```python
plt.figure(figsize=(15,8))
plt.scatter(Year, Japan_Waste, color='red', label='Japan')
plt.scatter(Year, Korea_Waste, color='green', label='South Korea')
plt.legend(loc='upper left')
plt.xlabel("Year")
plt.ylabel("x1000 Metric Ton")
plt.title("Total Municiple Waste Collected - Japan vs. South Korea")
plt.show()
```

Now we plot the same data in a bar chart. Amount line chart, scatter chart, and bar chart, bar chart is the best chart to compare two data because you can directly notice differences between data. 

```python
plt.figure(figsize=(15,8))
bar_width=0.4
x=np.arange(25)
plt.bar(x-0.2, Japan_Waste, width=bar_width, color='red', label='Japan')
plt.bar(x+0.2, Korea_Waste, width=bar_width, color='green', label='South Korea')
plt.legend(loc='upper left')
plt.xticks(x, Year)
plt.xlabel("Year")
plt.ylabel("x1000 Metric Ton")
plt.title("Total Municiple Waste Collected - Japan vs. South Korea")
plt.show()
```

At the end, we plot a box plot to look at the data. We can see that there are many outliners at the South Korea data. The box plot isn't telling us much information. Normally box plot is good for investigating new data set and see how the data distribute. 

```python
data=[Japan_Waste, Korea_Waste]

mpl_fig = plt.figure(figsize=(5,5))
ax = mpl_fig.add_subplot(111)

ax.boxplot(data, widths=0.5)
plt.xticks([1,2],['Japan', 'South Korea'])
plt.ylabel("x1000 Metric Ton")
plt.show()
```
