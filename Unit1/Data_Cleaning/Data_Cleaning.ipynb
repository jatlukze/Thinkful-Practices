```python
import numpy as np
import pandas as pd
import fuzzywuzzy
from fuzzywuzzy import process
import re
```

```python
df = pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
# According to above dataframe, we can see that the COST column have a dollar sign in front of each value.
# We want to delete this dollar sign here. 

# for cost in df['COST (£) charged to Wellcome (inc VAT when charged)']:
#        cost = re.sub('£','',str(cost))

df["COST (£) charged to Wellcome (inc VAT when charged)"] = df["COST (£) charged to Wellcome (inc VAT when charged)"].str.replace('£', '')

df["COST (£) charged to Wellcome (inc VAT when charged)"] = df["COST (£) charged to Wellcome (inc VAT when charged)"].str.replace('$', '')

df.head()
```

```python
df = df.sort_values('Journal title') # sort Journal title alphabetically

df['Journal title'] = df['Journal title'].str.strip() # strip white spaces

df['Journal title'] = df['Journal title'].str.lower() # Capitalized the title

df['Journal title'] = df['Journal title'].str.replace(' +', ' ') # remove extra whitespace within string

df['Journal title'] = df['Journal title'].str.replace('&', 'and') # replace & with and 

df['Journal title'] = df['Journal title'].str.replace('-', '') # remove -

df['Journal title'] = df['Journal title'].str.replace(':', '') # remove :

df['Journal title'] = df['Journal title'].str.replace('.', '') # remove .

df['Journal title'] = df['Journal title'].str.replace(',', '') # remove ,

df['Journal title'] = df['Journal title'].str.replace('(', '') # remove (

df['Journal title'] = df['Journal title'].str.replace(')', '') # remove )

journals = df['Journal title'].unique() # find all unique titles

print(journals)

print(len(journals))

df.head()

```
